## 3. Developing a Basic Chatbot Interface

We will start by creating a simple function that can answer questions based on the `taxila_knowledge_base`. This function will look for keywords in the user's question and return the corresponding information from our knowledge base.

```python
def answer_question_from_kb(question, kb):
    question = question.lower()

    if "history" in question:
        return kb.get("history", "I don't have information about the history of Taxila.")
    elif "location" in question:
        return kb.get("location", "I don't have information about the location of Taxila.")
    elif "cultural importance" in question or "importance" in question:
        return kb.get("cultural_importance", "I don't have information about the cultural importance of Taxila.")
    elif "features" in question or "major features" in question:
        features = kb.get("major_features", [])
        if features:
            return "Some major features of Taxila include: " + ", ".join(features) + "."
        else:
            return "I don't have information about the major features of Taxila."
    elif "architecture" in question:
        return kb.get("architecture", "I don't have information about the architecture of Taxila.")
    elif "visitor information" in question or "visit" in question:
        return kb.get("visitor_information", "I don't have information for visitors to Taxila.")
    elif "interesting facts" in question or "facts" in question:
        facts = kb.get("interesting_facts", [])
        if facts:
            return "Here are some interesting facts about Taxila: " + " ".join(facts) + "."
        else:
            return "I don't have any interesting facts about Taxila."
    elif "past appearance" in question or "look like in the past" in question:
        return kb.get("past_appearance", "I don't have details on what Taxila looked like in the past.")
    elif "name" in question:
        return kb.get("name", "I'm not sure of the name.")
    else:
        return "I'm sorry, I don't have an answer to that specific question in my knowledge base. Please try asking about its history, location, features, or architecture."

print("Chatbot function `answer_question_from_kb` defined.")
```

## 5. Enhancing the Chatbot with Gemini

Now we will create a function that uses the Gemini model to process user questions. This function will leverage the `taxila_knowledge_base` by providing it as context to Gemini, allowing the model to generate more intelligent and contextually rich answers.

```python
def get_gemini_response(question, knowledge_base):
    model = genai.GenerativeModel('gemini-pro')

    # Convert knowledge base to a readable string for the LLM
    kb_text = "Here is some information about Taxila:\n"
    for key, value in knowledge_base.items():
        if isinstance(value, list):
            kb_text += f"{key.replace('_', ' ').title()}: {', '.join(value)}\n"
        else:
            kb_text += f"{key.replace('_', ' ').title()}: {value}\n"

    prompt = f"""You are a helpful AI assistant providing information about the ancient site of Taxila.
Use the following knowledge base to answer questions about Taxila.
If the question cannot be answered from the provided knowledge base, state that you don't have enough information.
Do not make up information.

Knowledge Base:
{kb_text}

User Question: {question}

AI Answer:"""

    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"An error occurred while generating a response: {e}"

print("Gemini-enhanced chatbot function `get_gemini_response` defined.")
```

## 7. Creating a Streamlit User Interface

To provide an interactive demo, we will create a Streamlit application. This will combine our Gemini-powered chatbot and the Stable Diffusion image generation into a single user-friendly interface. We'll start by installing Streamlit.

### Streamlit Application Code

Now, we will write the Streamlit application code. This code will:
1.  Take a user's question.
2.  Use the `get_gemini_response` function to get a text response.
3.  Use the `generate_image` function to create an image based on the chatbot's response.
4.  Display both the text response and the generated image.

Since Colab notebooks don't directly run Streamlit apps in the output cell, we'll write the code to a file (`app.py`) and then provide instructions to run it using `streamlit run`.

```python
%%writefile app.py

import streamlit as st
import google.generativeai as genai
import torch
from diffusers import StableDiffusionPipeline
from google.colab import userdata
import matplotlib.pyplot as plt
from PIL import Image
import io
import base64

# --- 1. Knowledge Base for Taxila (copied from cell 3c9ac0f8) ---
taxila_knowledge_base = {
    "name": "Taxila",
    "location": "Punjab, Pakistan, near the modern city of Rawalpindi.",
    "history": "An ancient city and archaeological site dating back to the Gandhara period, with evidence of human habitation since the Neolithic era. It was a significant center of Buddhist learning and culture.",
    "cultural_importance": "A UNESCO World Heritage Site, renowned for its archaeological remains that represent various periods of South Asian history, including Persian, Greek, and Buddhist influences. It was a crossroads of cultures and a center for learning, especially for Buddhism.",
    "major_features": [
        "Bhir Mound (earliest settlement)",
        "Sirkap (Greek city built by Demetrius I of Bactria)",
        "Sirsukh (Kushan city)",
        "Dharmarajika Stupa and Monastery (important Buddhist complex)",
        "Mohra Muradu Monastery",
        "Jaulian Monastery"
    ],
    "architecture": "Features a blend of architectural styles, including Hellenistic, Persian, and indigenous South Asian elements, particularly evident in its stupas, monasteries, and city planning.",
    "visitor_information": "Taxila Museum houses a large collection of artifacts from the site. Accessible by road from Islamabad/Rawalpindi. Best visited during cooler months.",
    "interesting_facts": [
        "It is believed to be the site of one's earliest universities.",
        "Alexander the Great reportedly visited Taxila in 326 BC.",
        "It was an important trading center on the Silk Road."
    ],
    "past_appearance": "In its prime, Taxila would have been a bustling urban center with grand stupas, elaborate monasteries, well-planned cities like Sirkap, and a vibrant intellectual atmosphere, reflecting a blend of Hellenistic and Indic architectural styles with intricate carvings and sculptures."
}

# --- 2. Configure Gemini API (adapted from cell d9ac37d6) ---
@st.cache_resource
def configure_gemini():
    try:
        GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
        if not GEMINI_API_KEY:
            st.error("GEMINI_API_KEY not found in Colab Secrets. Please add it.")
            return None
        genai.configure(api_key=GEMINI_API_KEY)
        return True
    except Exception as e:
        st.error(f"Error configuring Gemini API: {e}. Please ensure your Gemini API key is added to Colab's Secrets Manager with the name 'GEMINI_API_KEY'.")
        return None

# --- 3. Gemini-enhanced Chatbot Function (copied from cell 2b6e96b7) ---
@st.cache_data
def get_gemini_response(question, knowledge_base):
    if not configure_gemini():
        return "Gemini API not configured. Please set your API key."

    model = genai.GenerativeModel('gemini-pro')

    kb_text = "Here is some information about Taxila:\n"
    for key, value in knowledge_base.items():
        if isinstance(value, list):
            kb_text += f"{key.replace('_', ' ').title()}: {', '.join(value)}\n"
        else:
            kb_text += f"{key.replace('_', ' ').title()}: {value}\n"

    prompt = f"""You are a helpful AI assistant providing information about the ancient site of Taxila.
Use the following knowledge base to answer questions about Taxila.
If the question cannot be answered from the provided knowledge base, state that you don't have enough information.
Do not make up information.

Knowledge Base:
{kb_text}

User Question: {question}

AI Answer:"""

    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"An error occurred while generating a response from Gemini: {e}"

# --- 4. Configure Stable Diffusion Pipeline (adapted from a previous cell) ---
@st.cache_resource
def load_stable_diffusion_pipeline():
    try:
        HF_TOKEN = userdata.get('HF_TOKEN')
        if not HF_TOKEN:
            st.error("HF_TOKEN not found in Colab Secrets. Please ensure it's set.")
            return None

        pipe = StableDiffusionPipeline.from_pretrained(
            "runwayml/stable-diffusion-v1-5",
            torch_dtype=torch.float16,
            use_auth_token=HF_TOKEN
        )

        if torch.cuda.is_available():
            pipe = pipe.to("cuda")
        else:
            pipe = pipe.to("cpu")
        return pipe
    except Exception as e:
        st.error(f"Error loading Stable Diffusion pipeline: {e}. Please ensure your HF_TOKEN is correctly set in Colab Secrets and you have accepted the model license on Hugging Face (e.g., runwayml/stable-diffusion-v1-5).")
        return None

# --- 5. Image Generation Function (adapted from a previous cell) ---
@st.cache_data
def generate_image(prompt, pipe):
    if pipe is None:
        return None

    st.info(f"Generating image for prompt: '{prompt}'...")
    try:
        with torch.autocast("cuda"):
            image = pipe(prompt).images[0]
        return image
    except Exception as e:
        st.error(f"Error generating image: {e}. This might happen if the GPU disconnects or for other reasons. Try refreshing or restarting the app.")
        return None

# --- Streamlit App Layout ---
st.set_page_config(layout="wide")
st.title("🏛️ Heritage Site AI Chatbot & Image Generator (Taxila)")
st.markdown("Ask questions about **Taxila** and get an AI-generated image based on the response!")

# Initialize chat history
if "messages" not in st.session_state:
    st.session_state.messages = []

# Load models once
with st.spinner('Loading Gemini and Stable Diffusion models...'):
    gemini_configured = configure_gemini()
    sd_pipe = load_stable_diffusion_pipeline()

# Display chat messages from history on app rerun
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])
        if "image" in message and message["image"] is not None:
            st.image(message["image"], caption=message["content"], use_column_width=True)

# React to user input
if prompt := st.chat_input("What would you like to know about Taxila?"):
    # Display user message in chat message container
    with st.chat_message("user"):
        st.markdown(prompt)
    # Add user message to chat history
    st.session_state.messages.append({"role": "user", "content": prompt})

    with st.chat_message("assistant"):
        with st.spinner("Getting answer from chatbot..."):
            response_text = get_gemini_response(prompt, taxila_knowledge_base)
            st.markdown(response_text)

        image_generated = None
        if sd_pipe is not None:
            with st.spinner("Generating image..."):
                # Generate image based on the chatbot's text response
                # Create a concise prompt for image generation from the response
                image_prompt = f"Ancient site of Taxila, {response_text[:150].strip()}, highly detailed, historical artwork, photorealistic"
                image_generated = generate_image(image_prompt, sd_pipe)
                if image_generated:
                    st.image(image_generated, caption=response_text, use_column_width=True)
        elif sd_pipe is None:
            st.warning("Image generation is disabled due to Stable Diffusion pipeline loading issues.")

    # Add assistant response and image to chat history
    st.session_state.messages.append({"role": "assistant", "content": response_text, "image": image_generated})

```

### Loading the Stable Diffusion Pipeline

We will load a Stable Diffusion model from Hugging Face. Ensure you have accepted the model's terms of use on Hugging Face (e.g., `runwayml/stable-diffusion-v1-5`). We will use the `HF_TOKEN` stored earlier to authenticate.

```python
import torch
from diffusers import StableDiffusionPipeline
from google.colab import userdata

try:
    HF_TOKEN = userdata.get('HF_TOKEN')
    if not HF_TOKEN:
        raise ValueError("HF_TOKEN not found in Colab Secrets. Please ensure it's set.")

    # Load the Stable Diffusion pipeline
    # Using 'runwayml/stable-diffusion-v1-5' as a common starting point.
    # You might need to accept its license on Hugging Face website first.
    pipe = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16,
        use_auth_token=HF_TOKEN
    )

    # Move the pipeline to GPU if available
    if torch.cuda.is_available():
        pipe = pipe.to("cuda")
        print("Stable Diffusion pipeline loaded and moved to GPU.")
    else:
        pipe = pipe.to("cpu")
        print("Stable Diffusion pipeline loaded to CPU (GPU not available).")

except Exception as e:
    print(f"Error loading Stable Diffusion pipeline: {e}")
    print("Please ensure your HF_TOKEN is correctly set in Colab Secrets and you have accepted the model license on Hugging Face.")
```

### Function for Image Generation

Now, let's create a function that takes a text prompt and generates an image using the loaded Stable Diffusion pipeline.

```python
import matplotlib.pyplot as plt
from PIL import Image
import io
import base64

def generate_image(prompt):
    if 'pipe' not in globals():
        return "Error: Stable Diffusion pipeline not loaded. Please run the previous cells."

    print(f"Generating image for prompt: '{prompt}'...")
    with torch.autocast("cuda"):
        image = pipe(prompt).images[0]

    # Display the image (optional, for notebook preview)
    plt.imshow(image)
    plt.title(prompt[:50] + "...")
    plt.axis('off')
    plt.show()

    # Save image to a BytesIO object and encode to base64 for potential future use (e.g., web display)
    img_byte_arr = io.BytesIO()
    image.save(img_byte_arr, format='PNG')
    img_byte_arr = img_byte_arr.getvalue()
    base64_encoded_image = base64.b64encode(img_byte_arr).decode('utf-8')

    return {"image": image, "base64_encoded": base64_encoded_image}

print("Image generation function `generate_image` defined.")
```
